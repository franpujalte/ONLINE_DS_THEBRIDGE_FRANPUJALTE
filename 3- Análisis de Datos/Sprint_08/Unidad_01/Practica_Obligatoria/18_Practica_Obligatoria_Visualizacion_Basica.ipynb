1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)

```python
ciudades = df.loc[df.population_total.nlargest(5).index, ["city","population_total"]]
nombres_ciudades = ["Los Angeles", "San Diego", "Sacramento", "Anaheim", "Santa Ana"]

plt.figure(figsize = (12,5))
plt.bar(nombres_ciudades, ciudades["population_total"]/1000, color = ("blue", "gray", "red", "olive", "brown"), alpha = 0.5)

#para poner los valores y de cada barra
for i, ciudad in enumerate(ciudades.index):
    poblacion = ciudades.loc[ciudad,"population_total"]/1000
    plt.text(nombres_ciudades[i], poblacion + 120, f"{poblacion}", ha = "center")

plt.ylim([0,5000])
plt.ylabel("Habitantes (miles)")
plt.xlabel("Ciudades")
plt.title("Ciudades ms pobladas de California")
plt.show()
```

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)

```python
plt.figure(figsize = (10,5))
plt.hist(df["area_total_km2"], color = "steelblue", bins = 50, label= "Hist Area", alpha = 0.9)

plt.xlabel("Superficie($km^2$)", fontname = "Bradley Hand ITC", fontsize = 18)
plt.ylabel("Nmero de ciudades", fontname = "Bradley Hand ITC", fontsize = 18)
plt.legend()
plt.show()
```

3. El scatter de rea de agua de un ciudad con su poblacion.

```python
plt.figure(figsize = (12,6))
plt.scatter(df["area_water_km2"], df["population_total"]/1000, color="green", alpha = 0.5)

props_flecha = {
    "arrowstyle": "<->",
}

outlier_1_y = df["population_total"].nlargest(1)
outlier_2_y = df["population_total"].nlargest(2).iloc[1]
outlier_1_x = df["area_water_km2"].nlargest(2).iloc[1]
outlier_2_x = df["area_water_km2"].nlargest(1)

plt.annotate("Outliers", xy = (outlier_1_x,outlier_1_y/1000), xytext = (50,3000),\
             arrowprops = props_flecha)
plt.annotate("Outliers", xy=(outlier_2_x, outlier_2_y/1000), xytext=(50, 3000),
             arrowprops= props_flecha)

plt.xlabel("rea de agua en la ciudad ($km^2$)")
plt.ylabel("Poblacin (miles de habitantes)")
plt.title("Poblacin por kilmetro cuadrado de agua")

plt.show()
```

4. El boxplot de la altura de las ciudades.

```python
plt.figure(figsize=(12,5))

plt.boxplot(df["elevation_m"],
            vert=False, 
            patch_artist=True, 
            boxprops=dict(facecolor="red"),
            medianprops=dict(color="black")) 


plt.xlabel("Altitud (m)")
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
```
